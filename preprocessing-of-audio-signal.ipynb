```python
#initiating all individual features

feature_amplitude_envelope=[]
feature_zero_crossing_rate=[]
feature_energy=[]
feature_spectral_centroid=[]
feature_spectral_contrast=[]
feature_mfccs=[]
feature_spectral_rolloff=[]




    
```

```python

for file in data:

    # Load the audio file
    audio_file = file.audio_path
    audio, sr = librosa.load(audio_file, sr=None)

    # Resampling
    target_sr = 44100  # Desired sampling rate
    resampled_audio = librosa.resample(audio, orig_sr=sr, target_sr=target_sr)

    # Normalization
    normalized_audio = resampled_audio / np.max(np.abs(resampled_audio))

    # Feature Extraction
    amplitude_envelope = np.abs(normalized_audio)
    
    feature_amplitude_envelope.append([np.mean(amplitude_envelope),np.var(amplitude_envelope), np.std(amplitude_envelope)])
    
    
    
    zero_crossing_rate = librosa.zero_crossings(normalized_audio, pad=False).sum()
    
    feature_zero_crossing_rate.append([zero_crossing_rate])
    
    
    
    energy = np.sum(normalized_audio ** 2)
    
    feature_energy.append([energy])
    
    
    
    spectral_centroid = librosa.feature.spectral_centroid(y=normalized_audio, sr=target_sr)
    
    feature_spectral_centroid.append([np.mean(spectral_centroid),np.var(spectral_centroid),np.std(spectral_centroid)])
    
    
    
    spectral_contrast = librosa.feature.spectral_contrast(y=normalized_audio, sr=target_sr)
    feature_spectral_contrast.append([np.mean(spectral_contrast),np.var(spectral_contrast),np.std(spectral_contrast)])
    
    mfccs = librosa.feature.mfcc(y=normalized_audio, sr=target_sr)
    feature_mfccs.append([np.mean(mfccs),np.var(mfccs),np.std(mfccs)])
    spectral_rolloff = librosa.feature.spectral_rolloff(y=normalized_audio, sr=target_sr)
    feature_spectral_rolloff.append([np.mean(spectral_rolloff),np.var(spectral_rolloff),np.std(spectral_rolloff)])
    
 

   
```

```python
target_data=[]
for file in data:
    target_data.append(file.stroke_name)
target_data=np.array(target_data)
len(target_data)
```

```python
feature_amplitude_envelope=np.array(feature_amplitude_envelope)
feature_zero_crossing_rate=np.array(feature_zero_crossing_rate)
feature_energy=np.array(feature_energy)
feature_spectral_centroid=np.array(feature_spectral_centroid)
feature_spectral_contrast=np.array(feature_spectral_contrast)
feature_mfccs=np.array(feature_mfccs)
feature_spectral_rolloff=np.array(feature_spectral_rolloff)
```

```python
final_feature_data=np.concatenate((feature_amplitude_envelope,feature_spectral_centroid,feature_spectral_contrast,feature_mfccs,feature_spectral_rolloff),axis=1)
```
